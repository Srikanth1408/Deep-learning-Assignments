```python
def entropy(p):
    a=0
    for i in p:
        a=a+(-i*np.log2(i))
    return a
```

```python
def kl_divergence(p,q):
    a=0
    for i,j in zip(p,q):
        a=a+i*(np.log2(i)-np.log2(j))
    return a
```

```python
def kl_divergence1(q,p):
    a=0
    for i,j in zip(p,q):
        a=a+i*(np.log2(i)-np.log2(j))
    return a
```

```python
def cross_entropy(p,q):
    a1=entropy(p)
    a2=kl_divergence(p,q)
    return a1+a2
```

```python
def cross_entropy1(p,q):
    a=0
    for i,j in zip(p,q):
        a=a+(i*np.log2(j))
    return -a
```

```python
def JS_divergence(p,q):
    m=[]
    for i,j in zip(p,q):
        m.append((i+j)/2)
    return kl_divergence(p,m)+kl_divergence(q,m)
```

```python
def wasserstein(joint_probability,r,s):
  x=np.sum(joint_probability,axis=1)
  y=np.sum(joint_probability,axis=0)
  print(x,y)
  if (np.allclose(x,r) and np.allclose(y,s) ):
    print("True it belongs")
  else:
    print("False It doesn't belong")
```

```python
def convolution(pic , kernel):

    s_hgt = pic.shape[0]
    s_wth = pic.shape[1]
    s_dth = 3
    ar = np.zeros((s_hgt,s_wth,s_dth))
    
    for x in range(s_dth):
        
        pic_ = pic[:,:,x]
        kernel1 = kernel[:,:,x]
                                            #flipping kernel as either of kernel or image should be flipped                                                
        kernel1 = np.flip(kernel1)
        kernel1h, kernel1w = kernel1.shape

                                            #padding is done to take care of proper convolution
        pic_pad = np.pad(pic_, kernel1h - 1 , mode = 'edge')
        pic_h , pic_w = pic_pad.shape

        m = pic_h - kernel1h + 1
        n = pic_w - kernel1w + 1
        convl_pic = np.zeros((m,n))

                                              # Move through the row and coulum for 258 size

        for i in range(m):
            for j  in range(n):
                                              # get the subarray for multiplying and further adding

                sub_array = pic_pad[i :i + kernel1h , j:j + kernel1w]
                convl_pic[i,j] = np.sum(sub_array * kernel1)

        convl_pic = convl_pic[1:257,1:257]
        ar[:,:,x] = ar[:,:,x] + convl_pic
        
    sum = np.zeros((s_hgt, s_wth));
    sum = ar[:,:,0]+ar[:,:,1]+ar[:,:,2]
    return sum
```

```python
def correlation(pic , kernel):

    s_hgt = pic.shape[0]
    s_wth = pic.shape[1]
    s_dth = 3
    ar = np.zeros((s_hgt,s_wth,s_dth))
    
    for x in range(s_dth):
        
        pic_ = pic[:,:,x]
        kernel1 = kernel[:,:,x]
                                            #flipping kernel is not done with correlation operation, so it is removed                                                
        # kernel1 = np.flip(kernel1)
        kernel1h, kernel1w = kernel1.shape

                                            #padding is done to take care of proper convolution
        pic_pad = np.pad(pic_, kernel1h - 1 , mode = 'edge')
        pic_h , pic_w = pic_pad.shape

        m = pic_h - kernel1h + 1
        n = pic_w - kernel1w + 1
        convl_pic = np.zeros((m,n))

                                              # Move through the row and coulum for 258 size

        for i in range(m):
            for j  in range(n):
                                              # get the subarray for multiplying and further adding

                sub_array = pic_pad[i :i + kernel1h , j:j + kernel1w]
                convl_pic[i,j] = np.sum(sub_array * kernel1)

        convl_pic = convl_pic[1:257,1:257]
        ar[:,:,x] = ar[:,:,x] + convl_pic
        
    sum = np.zeros((s_hgt, s_wth));
    sum = ar[:,:,0]+ar[:,:,1]+ar[:,:,2]
    return sum
```

Convolution in commutative whereas correlation is not.

i)
The distances in the neighbourhood are preserved in two dimensions using    t-sne. However, in some circumstances, it is not possible to save all of the neighbourhoods. This is referred to as a crowding problem.
In ten dimensions, for example, it is conceivable to have 11 datapoints that are mutually equidistant, which is impossible to do in two dimensions.

ii)
The large tails of the student t-distribution provide more distant neighbours more flexibility in where they are put, which helps to relieve the crowding problem that plagued the original SNE algorithm. In t-SNE, the "clumps" in SNE became "islands."

iii)
The scale student's t distribution probability function in low dimensionality probability space is a key change in t-SNE compared to SNE because t-SNE employs t-distribution to compute the conditional probability in low dimensionality probability space, thus the data projection has a broader range. In addition, T-SNE has a lower error than SNE.
