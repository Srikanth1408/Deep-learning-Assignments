```python
import numpy as np
def normball(p):
    x1=np.array(x)
    y1=np.power((1-np.power(abs(x1),p)),1/p)
    y2=-np.power((1-np.power(abs(x1),p)),1/p)
    plt.plot(x,y1)
    plt.plot(x,y2)
plt.subplot(1,2,1)
for i in range(1,5,1):
    normball(i)
plt.subplot(1,2,2)
for i in range(1,10,1):
    normball(i/10)

```

```python
import tifffile as tiff
def normalized_hist(pic):
    a = tiff.imread(pic)
    print(a)
    
    pixel_val = np.unique(a)
    count=[]
    for i in pixel_val:
        c = np.count_nonzero(a == i)
        count.append(c)
    sq=np.sqrt(sum(count))
    print("The image size is of ",sq,"*",sq,"pixels")
    for i in range(len(count)):
        count[i]=count[i]/(sq*sq)
    no_of_pixels = count
    pixel_val.tolist()
    plt.figure(figsize=(15, 9))
    plt.bar(pixel_val, no_of_pixels)
    print(sum(count))

normalized_hist('5.1.09.tiff')
```

```python
def kl_divergence(p,q):
    a=0
    for i,j in zip(p,q):
        a=a+i*(np.log2(i)-np.log2(j))
    return a
```

```python
def kl_divergence1(q,p):
    a=0
    for i,j in zip(p,q):
        a=a+i*(np.log2(i)-np.log2(j))
    return a
```

```python
p=[0.1,0.8,0.05,0.05]
q=[0.7,0.23,0.05,0.02]
print(kl_divergence(p,q),kl_divergence1(p,q))
```
